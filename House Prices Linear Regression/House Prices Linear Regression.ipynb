The reason for the creation of a correlation matrix is to locate <b>multicollinearity</b>. In simplest terms, multicollinearity occurs when independent variables are correlated. This is an issue because the goal of regression analyses is to isolate the relationship between each independent variable and dependent variable. 

The regression coefficient from the regression analyses represents the mean change in the dependent variable for each one-unit change in the independent variable when all the other independent variables are held constant. 

One way to "fix" multicollinearity is to calculate Variance Inflation Factor (VIF), which detects multicollinearity in a regression analysis, and dropping the columns with VIF > 10 <b>after</b> training the dataset. The VIF estimates how much the variance of a regression coefficient is inflated due to multicollinearity in the model. A VIF = 1 means no multicollinearity. Usually, 10 is the upper threshold, but it depends on the statistician.

# Linear Regression Modeling 

```python
# scikit-learn has a function to partition data into train and test
from sklearn.model_selection import train_test_split
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
# isolated the normally distributed log Price (inde_var_log) and feature engineered dataframe (dumVar_num_df)
x_train, x_test, y_train, y_test = train_test_split(dumVar_num_df, inde_var_log, test_size = 0.3, random_state = 0)

print("x_train", x_train.shape)
print("x_test", x_test.shape)
print("y_train", y_train.shape)
print("y_test", y_test.shape)
```

```python
# statsmodel is a Python module that provides classes and functions for the estimation of many different statistical 
# models, as well as for conducting statistical tests, and statistical data exploration.
import statsmodels.api as sm
# Create linear regression using OLS method
lR_model = sm.OLS(y_train, x_train).fit()
```

OLS means Ordinary Least Squares, and it is a commonly used technique in econometrics to reduce error. Error refers to the difference between an observed and a predicted point; essentially, the distance between a data point and the regression line. OLS computes the squares of the errors and then attempts to find a regession line through the sample data that minimizes the sum of the squared errors. Thus, a lower number means better predictor.

```python
# Show a summary of linear regression
lR_model.summary()
```

Root Mean Square Error (RMSE) is the standard deviation of the residuals, or errors. RMSE tells the analyst how concentrated the data points are around the line of best fit (regression line). Lower values of RMSE indicate better fit.

Calculating RMSE: 
https://statweb.stanford.edu/~susan/courses/s60/split/node60.html

Translation:
1. Find the differences between line of best fit and data point
2. Square the differences; gets rid of negative numbers
3. Find the mean of the squared differences
4. Square root the sum of squared differences

```python
# Use a function to save the rmse calculations algorithm
def RMSError(predicted, actual):
    differences = predicted - actual
    differences_squared = differences ** 2
    differences_squared_mean = differences_squared.mean()
    rmse_value = np.sqrt(differences_squared_mean)
    return rmse_value 
```

```python
# Create a pandas dataframe to present the findings
cols_names = ["Model", "R-Squared Value", "Adjusted R-Squared Value", "RMSError"] #names of columns
df_col_stats = pd.DataFrame(columns = cols_names)
# fit the regression on the test dataset; needed to calculate RMSError value
y_predict1 = lR_model.predict(x_test)
```

```python
# insert a row into the df_col_stats dataframe with the calculated statistics
df_col_stats.loc[0] = ["Linear Regression Statistics", lR_model.rsquared, lR_model.rsquared_adj, 
                        RMSError(y_predict1, y_test)]
df_col_stats 
```

At the simplest level, r-squared evaluates the scatter of the data points around the fitted regression line. R-Squared is between 0 - 100% (0 - 1) and a higher percentage means the better the regression model fits the observations. 

Adjusted R-Squared is similar. The main difference: r-squared assumes every independent variable explains the variation in the dependent variable. However, the adjusted r-squared tells the percentage of variation explained by only the independent variables that actually affect the dependent variable. Thus, adjusted r-squared is always lower than the r-squared 

# New Linear Regression Model with Constant

```python
# add a column of 1s to the dataset; acts as the intercept
df_constant = sm.add_constant(dumVar_num_df)
```

```python
# isolated the normally distributed log Price (inde_var_log) and feature engineered dataframe (dumVar_num_df)
x_train2, x_test2, y_train2, y_test2 = train_test_split(df_constant, inde_var_log, test_size = 0.3, random_state = 0)
```

```python
# Create linear regression using OLS method
model2 = sm.OLS(y_train2, x_train2).fit()
model2.summary()
```

```python
# fit the regression on the test dataset; needed to calculate RMSError value
y_predict2 = model2.predict(x_test2)
#names of columns
cols_names2 = ["Model", "R-Squared Value", "Adjusted R-Squared Value", "RMSError"]
# create a dataframe with columns specified in cols_names2
df_col_stats2 = pd.DataFrame(columns = cols_names2)
# insert a row into the df_col_stats dataframe with the calculated statistics
df_col_stats2.loc[0] = ["Linear Regression Statistics", model2.rsquared, model2.rsquared_adj, 
                        RMSError(y_predict2, y_test2)]

df_col_stats2 
```

# Variance Inflation Factor (VIF)

```python
# store the column labels
cols_names3 = x_train2.columns
# loop amount equals length of cols_names2

# store vif outputs; need for removal
Names_vif = pd.DataFrame(columns = ["name", "vifs"])

for i in np.arange(0, len(cols_names3)):
    # list of independent variable names in a list
    indVar = list(cols_names3)
    # delete a column per iteration
    depVar = indVar.pop(i)
    # compare each independent variable to the dependent variable one column at a time... until 0 
    modCalc = sm.OLS(x_train2[depVar], x_train2[indVar])
    # regression line fitting 
    resid = modCalc.fit()
    # calculate VIF
    vif = 1 / (1 - resid.rsquared)
    Names_vif.loc[i] = [depVar, round(vif, 3)]   
```

Remove all variables with VIF above 10.

```python
# put all variables in "name" into a list
name_all = Names_vif["name"].tolist()
# put all variables in vifs into a list
vif_all = Names_vif["vifs"].tolist()

# empty list to append desired results
vif_10 = []

# search through the length of the list, and if the boolean is True, then append the name indexed at i to empty list
for i in range(len(name_all) - 1):
    if vif_all[i] < 11:
        vif_10.append(name_all[i])
    else:
        pass

# create new dataset called x_train3 with only the desired columns        
x_train3 = x_train2[vif_10]
x_train3.head()
```

```python
# Create linear regression using OLS method
model3 = sm.OLS(y_train2, x_train3).fit()
```

```python
# new test dataset with only the desired columns
x_test3 = x_test2[vif_10]
```

```python
# fit the regression on the test dataset; needed to calculate RMSError value
y_predict3 = model3.predict(x_test3)
#names of columns
cols_names3 = ["Model", "R-Squared Value", "Adjusted R-Squared Value", "RMSError"]
# create a dataframe with columns specified in cols_names3
df_col_stats3 = pd.DataFrame(columns = cols_names3)
# insert a row into the df_col_stats dataframe with the calculated statistics
df_col_stats3.loc[0] = ["Linear Regression Statistics", model3.rsquared, model3.rsquared_adj, 
                        RMSError(y_predict3, y_test2)]

df_col_stats3 
```

The initial linear regression model was in the form y = ax. The new model utilizes the form y = ax + b. When the intercept was added, the root mean square error dropped from 0.40 to 0.16, which was a significant improvement. The lower root mean square error meant the model was a better predictor.

<b> How to interpret results? </b>

Model 3 shows the coefficients, which are the numerical independent variables impact on the dependent variable. Standard errors refer to the standard deviation of the coefficient's sampling distribution. Subsequently, the t-values show how statistically significant each coefficient is in the model. Moreover, P>|t| refers to the relationship between the independent and dependent variables. Generally, the P signifies the null hypothesis that the coefficient is equal to 0; the pre-selected alpha value is 0.05. Therefore, any value less than 0.05 shows a significant relationship between the variables. For example, if alpha is 0.05, then coefficients having a p-value of 0.05 or less would be statistically significant (i.e., null hypothesis is rejected, and say that the coefficient is significantly different from 0).
